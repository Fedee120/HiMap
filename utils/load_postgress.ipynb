```python
import psycopg2
import json
import os

# Define PostgreSQL connection parameters
POSTGRES_HOST = 'localhost'  # Adjust this to the address of your Docker host
POSTGRES_PORT = '5432'
POSTGRES_DB = 'wiki'
POSTGRES_USER = 'postgres'
POSTGRES_PASSWORD = 'yourpassword'

# Connect to the PostgreSQL database
conn = psycopg2.connect(
    dbname=POSTGRES_DB,
    user=POSTGRES_USER,
    password=POSTGRES_PASSWORD,
    host=POSTGRES_HOST,
    port=POSTGRES_PORT
)
cursor = conn.cursor()

# Define the directory containing your JSON files
json_directory = 'I:\HiMap\split_files'

# Set batch size
batch_size = 1000  # Adjust this value as needed
record_count = 0

# Iterate through each JSON file in the specified directory
for filename in os.listdir(json_directory):
    if filename.endswith('.json'):
        file_path = os.path.join(json_directory, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            print(f'Processing file: {filename}')  # Print the name of the file being processed
            records = json.load(file)
            for record in records:
                # Insert data into the database
                cursor.execute(
                    "INSERT INTO summaries (id, summary) VALUES (%s, %s) ON CONFLICT (id) DO NOTHING;",
                    (record['id'], record['summary'])
                )
                record_count += 1
                # Commit the transaction in batches
                if record_count % batch_size == 0:
                    conn.commit()
                    print(f'Committed {record_count} records.')

# Commit any remaining records
conn.commit()
print(f'Committed {record_count} records.')

# Close the database connection
cursor.close()
conn.close()
```
